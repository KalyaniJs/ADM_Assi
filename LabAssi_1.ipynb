```python
"""dataset from kaggle and perform following operations on given dataset
a.Description of data i.e. how many attributes are there
b.Tell about types of attributes
c.Statistical distribution of data- find mean, mode and median of data.
d.draw the box ploat for given data """
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset_path = "/content/netflix_movies_detailed_up_to_2025.csv"
df = pd.read_csv(dataset_path)
```

```python
#display first five row
print(df.head())
```

```python
# Define numerical columns
numerical_cols = ["release_year", "rating", "popularity", "vote_count", "vote_average", "budget", "revenue"]

# Compute mean, median, and mode
stats_summary = {
    "Attribute": [],
    "Mean": [],
    "Median": [],
    "Mode": []
}

for col in numerical_cols:
    stats_summary["Attribute"].append(col)
    stats_summary["Mean"].append(df[col].mean())
    stats_summary["Median"].append(df[col].median())
    stats_summary["Mode"].append(df[col].mode()[0])  # Taking the first mode value

# Convert to DataFrame
stats_df = pd.DataFrame(stats_summary)
print(stats_df)
```

```python
# Generate Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=45)
plt.title("Box Plot of Numerical Attributes in Netflix Movies Dataset")
plt.show()
```
